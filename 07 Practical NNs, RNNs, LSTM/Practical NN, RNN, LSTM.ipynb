```python
input = [3,4,5,6,7,6,7,5,6,7,8,9]
output = [3,4,5,56,6,7,8,9,0,0,7,4]
```

```python
list(zip(input,output))
```

```python
w = 0.0
b= 0.0
lr = .01
```

```python
for epoch in range(1000):
    total_loss = 0
    for x,y in zip(input,output):
        Y_pred = w*x+b
        error = Y_pred-y
        loss = error**2
        total_loss = total_loss + loss
        
        dw = 2*error*x
        db = 2*error
        
        w = w -lr*dw
        b = b - lr*db
        
    if epoch %10 ==0 :
        print(f"epoch {epoch} | Loss : {total_loss}  | w : {w} , b: {b}")
        
test_data = 7 
prediction = w*test_data + b 
print(prediction)   
```

```python
input = [1,2,3,4,5]
output = [2,4,6,8,10] 
```

```python
w1 , b1 = .1,0.0
w2,b2 = -0.1 , 0.0
w3,w4 = .1 , .1 
b3 = 0.0
lr = .01
```

```python
def relu(x):
    return max(0,x)
```

```python
def relu_derivative(x):
    return 1 if x>0 else 0 
```

```python
for epoch in range(100):
    total_loss = 0 
    for x,y in zip(input,output):
        z1 = w1*x + b1
        h1 = relu(z1)
        
        z2 = w2*x + b2
        h2 = relu(z2)
        
        Y_pred = w3*h1 + w4*h2 + b3
        
        error = Y_pred - y 
        loss  = error**2
        total_loss = total_loss + loss
        
        dl_dy = 2*error
        
        dw3 = dl_dy * h1
        dw4 = dl_dy * h2
        db3 = dl_dy
        
        dh1 = dl_dy * w3
        dh2 = dl_dy * w4
        
        dz1 = dh1* relu_derivative(z1)
        dz2 = dh2* relu_derivative(z2)
        
        dw1 = dz1 * x
        db1 = dz1
        
        dw2 = dz2 * x
        db2 = dz2
        
        w1 = w1 - lr*dw1
        w2 = w2 - lr*dw2
        w3 = w3 - lr*dw3
        w4 = w4 - lr*dw4
        
        
        b1 = b1 - lr*db1
        b2 = b2 - lr*db2
        b3 = b3 - lr*db3
        
    if epoch %10 ==0 :
        print(f"epoch {epoch} | Loss : {total_loss} ")
        
        
def prediction(x):
    z1 = w1*x + b1
    h1 = relu(z1)
        
    z2 = w2*x + b2
    h2 = relu(z2)
        
    y_output = w3*h1 + w4*h2 + b3
    return y_output     
                
prediction(9)     
```

```python
data = 'help'
char2idx = {'h':0 , 'e' : 1 , "l":2 , "p" :3}
idx2char = {v : k for k , v in char2idx.items()}
```

```python
char2idx
```

```python
idx2char
```

```python
input_seq = [char2idx[c] for c in data[:-1]]
target_seq = [char2idx[c] for c in data [1:]]
```

```python
target_seq
```

```python

input_seq
```

```python
x = tf.one_hot(input_seq,depth=4 )
x = tf.expand_dims(x , axis=0)
y = tf.expand_dims(target_seq , axis=0)
```

```python
x
```

```python
y
```

```python
model = tf.keras.Sequential([tf.keras.layers.LSTM(8,return_sequences=True , input_shape =(3,4) ) , 
                     tf.keras.layers.Dense(4,activation='linear')])
```

```python
model.compile(optimizer='adam' ,loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True))
```

```python
for epoch in range(201):
    model.train_on_batch(x,y)
    
    if epoch % 50  == 0 :
        output = model(x)
        print(output)
        pred = tf.argmax(output,axis = -1).numpy()[0]
        pred_char = "".join([idx2char[i] for i in pred] )
        print(f"epoch : {epoch} , prediction : {pred_char} , loss : {loss} ")
```
